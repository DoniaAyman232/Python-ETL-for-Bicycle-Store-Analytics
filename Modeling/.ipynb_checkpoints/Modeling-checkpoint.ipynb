```python
import pandas as pd
import os
output_path = '../Information Mart/orders_products_details/orders_products_details.csv'
os.makedirs(os.path.dirname(output_path), exist_ok=True)
details.to_csv(output_path, index=False)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

orders_products_details = pd.read_csv('../Information Mart/orders_products_details/orders_products_details.csv')

sales_per_category = orders_products_details.groupby('category_id')['list_price_y'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_per_category.index, y=sales_per_category.values)
plt.xlabel('Category ID')
plt.ylabel('Total Sales')
plt.title('Total Sales per Product Category')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(orders_products_details['latency days'].dropna(), bins=30, kde=True)
plt.xlabel('Latency Days')
plt.ylabel('Frequency')
plt.title('Distribution of Order Latencies')
plt.tight_layout()
plt.show()
```

```python
 
orders_products_details['order_date'] = pd.to_datetime(orders_products_details['order_date'])
sales_over_time = orders_products_details.groupby('order_date')['list_price_y'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=sales_over_time.index, y=sales_over_time.values)
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='quantity', y='list_price_y', data=orders_products_details)
plt.xlabel('Quantity')
plt.ylabel('List Price')
plt.title('Quantity vs. List Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='category_id', y='list_price_y', data=orders_products_details)
plt.xlabel('Category ID')
plt.ylabel('List Price')
plt.title('List Prices per Product Category')
plt.xticks(rotation=45)
plt.tight_layout()


plt.show()
```
